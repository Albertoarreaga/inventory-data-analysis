```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

data = {
    "producto_id": range(1, 21),
    "producto": [f"Producto_{i}" for i in range(1, 21)],
    "categoria": np.random.choice(["Materia Prima", "Producto Terminado"], 20),
    "inventario_sistema": np.random.randint(50, 500, 20),
    "inventario_fisico": np.random.randint(45, 520, 20),
    "entradas_mes": np.random.randint(10, 200, 20),
    "salidas_mes": np.random.randint(10, 200, 20),
    "costo_unitario": np.random.randint(50, 500, 20)
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.info()
```

```python
df["diferencia_inventario"] = df["inventario_fisico"] - df["inventario_sistema"]
df["error_porcentual"] = (df["diferencia_inventario"] / df["inventario_sistema"]) * 100
df["rotacion_inventario"] = df["salidas_mes"] / df["inventario_sistema"]
df["dias_inventario"] = 30 / df["rotacion_inventario"]
df["valor_inventario"] = df["inventario_sistema"] * df["costo_unitario"]

df.round(2)
```

```python
df.sort_values("error_porcentual", ascending=False).head()
```

```python
df.sort_values("rotacion_inventario").head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df["producto"], df["diferencia_inventario"])
plt.xticks(rotation=90)
plt.title("Diferencia entre Inventario Fsico y Sistema")
plt.xlabel("Producto")
plt.ylabel("Unidades")
plt.show()
```

El anlisis permiti identificar productos con diferencias relevantes entre inventario fsico y sistema, as como artculos con baja rotacin que representan riesgo de sobrestock. Los resultados sugieren oportunidades para mejorar controles de inventario, optimizar polticas de reposicin y reducir costos operativos mediante decisiones basadas en datos.
